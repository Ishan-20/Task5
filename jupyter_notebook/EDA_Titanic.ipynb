# Univariate Analysis

```python
df['Age'].hist(bins=30)
plt.title("Distributed of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
```

# Bivariate Analysis

```python
sns.countplot(x='Sex', hue='Survived', data=df)
sns.countplot(x='Pclass', hue='Survived',data=df)
```

# Correlation Heatmap

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df.drop(columns='Cabin', inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

# Observations

```python
Majority of survivors were female, most passengers were in 3rd class.
```

# Summary of Findings

```python
- 38% passengers survived
- Female passengers had a much higher survival rate than males
- First-class passengers had better survival chances
- Age and Fare show some correlation with survival
- Most missing data was in `Cabin`, so dropped; `Age` and `Embarked` were imputed
```
